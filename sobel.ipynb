```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image=plt.imread('img/show.jpg')
image.shape
plt.imshow(image)
```

```python
gray=rgb2gray(image)
plt.imshow(gray, cmap='gray')
```

```python
gx=np.array([np.array([-1,0,1]), np.array([-2,0,2]), np.array([-1,0,1])])
gy=np.array([np.array([1,2,1]), np.array([0,0,0]), np.array([-1,-2,-1])])
hHP=np.array([np.array([-1/16,-1/8,-1/16]), np.array([-1/8,3/4,-1/8]), np.array([-1/16,-1/8,-1/16])])
hMF=np.array([np.array([1/25,1/25,1/25,1/25,1/25]), np.array([1/25,1/25,1/25,1/25,1/25]),
              np.array([1/25,1/25,1/25,1/25,1/25]), np.array([1/25,1/25,1/25,1/25,1/25]), 
              np.array([1/25,1/25,1/25,1/25,1/25])])
```

```python
out_gx=ndimage.convolve(gray, gx, mode='reflect')
out_gy=ndimage.convolve(gray, gy, mode='reflect')
out_hHP=ndimage.convolve(gray, hHP, mode='reflect')
out_hMF=ndimage.convolve(gray, hMF, mode='reflect')
```

```python
plt.rcParams['figure.figsize']=(18,16)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(out_gx, cmap='gray')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(out_gy, cmap='gray')
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(out_hHP, cmap='gray')
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(out_hMF, cmap='gray')
```
